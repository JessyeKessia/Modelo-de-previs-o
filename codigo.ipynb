```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Caminho do CSV criado anteriormente
caminho_csv = "data/sorvete_temperatura_vendas.csv"

# 1. Ler os dados do CSV
df = pd.read_csv(caminho_csv)

print(df.columns)
```

```python
# 2. Preparar os dados
X = df[["Temperatura (C)"]]   # Feature
y = df["Vendas (unidades)"]   # Target
```

```python
# 3. Treinar o modelo de regressÃ£o linear
model = LinearRegression()
model.fit(X, y)
```

```python

# 4. Fazer a previsÃ£o de vendas com base na temperatura
df["Estoque Previsto"] = model.predict(X).round().astype(int)
```

```python
# 5. Mostrar a equaÃ§Ã£o da previsÃ£o
print(f"EquaÃ§Ã£o: vendas = {model.coef_[0]:.2f} * temperatura + {model.intercept_:.2f}")
print("\nPrevisÃµes de estoque por dia:\n")
print(df[["Data", "Temperatura (C)", "Estoque Previsto"]])
```

```python


# 6. Visualizar graficamente
plt.plot(df["Data"], df["Estoque Previsto"], marker='o', label="Estoque Previsto")
plt.plot(df["Data"], df["Vendas (unidades)"], marker='x', label="Vendas Reais")
plt.xticks(rotation=45)
plt.xlabel("Data")
plt.ylabel("Unidades de Sorvete")
plt.title("PrevisÃ£o de Estoque de Sorvete por Temperatura")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

```
